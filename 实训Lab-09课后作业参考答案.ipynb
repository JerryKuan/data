```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

# dataset = pd.read_csv('/home/jerry/GitRepository/data/data/iris.csv')
dataset = pd.DataFrame(load_iris().data)
dataset.columns = ['sepal length','sepal width','petal length','petal width']
dataset['Species'] = load_iris().target
dataset.describe()
```

```python
dataset.info()
```

```python
'''描述sepal length和类别之间的关系'''
g = sns.FacetGrid(dataset, col = 'Species')
g.map(plt.hist, 'sepal length', bins = 20)
```

```python
'''描述sepal length和sepal width,petal length和petal width之间的联系'''
dataset.plot(kind = 'scatter', x = 'sepal length', y = 'sepal width')
dataset.plot(kind = 'scatter', x = 'petal length', y = 'petal width')
```

```python
'''用不同颜色的点来表示数据所属的类别'''
sns.FacetGrid(dataset, hue = 'Species', size =  5).map(plt.scatter, 'sepal length', 'sepal width').add_legend()
```

```python
'''通过箱线图查看单个特征的分布'''
sns.boxplot(x = 'Species', y='petal length', data = dataset)
```

```python
'''在箱线图基础上添加散点图,jitter设置成True使散点图分开,不然会连成一条线'''
ax = sns.boxplot(x = 'Species', y = 'petal length', data = dataset)
ax = sns.stripplot(x = 'Species', y = 'petal length', data = dataset, jitter = True, edgecolor = 'gray')
```

```python
'''核密度图（单个变量）'''
sns.FacetGrid(dataset, hue="Species", size=6).map(sns.kdeplot, "petal length").add_legend()
```

```python
'''描述任意两个变量之间的关系'''
sns.pairplot(dataset, hue="Species", size=3)
```
